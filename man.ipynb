```python
import random
# reading the text file
words = open("hang.txt",  "r")
words = words.read()

def get_word():
    # split to turn to a list and return the a random choice
    word = random.choice(words.split())
    #print(word)
    return word


def play(word):
    word = random.choice(words.split())
    print("this is a", len(word), "letter word")
    word_tofill = "_" * len(word)
    # A list where the word will be appended to
    guessed = False
    guesses = []
    trial = 0
    # giving a limited amount of trials of five(5)
    print("Lets play hangman")
    print(trial)
    print(word_tofill)
    print("\n")
    
    while trial < 10:
    #   input of the user
        Input = input("guess a letter in the word: ")
    #   check if the user has already guess the word
        if Input in word:
            if Input in guesses:
                print("already used this word")
                trial += 1
    #       I need to use the keyword "eneumerate" to complete this project
            else:
#             on this occasion the user has inputed a corrrect guess
                print('correct guess, keep going')
#               now this turns the length of the underscore to a list
                word_list = list(word_tofill)
#               this part is one the most important part of the code 
#               making sure that his input is inserted in the right spot in the underscore
#               by using the indices a and a foor loop that gives a position for everything
                indices = [i for i, letter in enumerate(word) if letter == Input]
                for index in indices:
                    word_list[index] = Input
                word_tofill = "".join(word_list)
                print(word_tofill)
                    
                if "_" not in word_tofill:
                    guessed = True
#                 guesses.insert(word.index(Input), Input)
        else:
            print("incorrect guess")
            trial += 1
#         if len(guesses) == len(word):
#             print("your guesses were correct and the word is", word)
#             break
#         print(guesses)
    if guessed:
        print("congrats you guessed the word, you win")
    else:
        print("sorry you ran out of tries the word was", word,"maybenext time")

def main():
    word = get_word
    play(word)
    while input("play again (Y/N)").upper() == "Y":
        word = get_word
        play(word)

if __name__ =="__main__":
    main()
    
```
